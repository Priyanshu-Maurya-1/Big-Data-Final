```python
from config import client, db, collection
import requests
import json
from pymongo import MongoClient

def fetch_all_data(batch_size=100):
    all_data = []
    start = 0
    while True:
        print(f"Fetching records {start} to {start + batch_size}...")
        url = f"{API_URL}/trips?start={start}&end={start + batch_size}"
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"Error fetching data: {response.status_code}")
            break
        
        data = response.json()
        if not data:  
            break
        
        all_data.extend(data)
        start += batch_size
    return all_data

def save_data_locally(data, filename="bike_trips.json"):
    with open(filename, "w") as file:
        json.dump(data, file)
    print(f"Data saved locally to {filename}")

def insert_data_into_mongodb(data):
    if data:
        collection.insert_many(data)
        print(f"Inserted {len(data)} records into MongoDB!")
    else:
        print("No data to insert into MongoDB.")

if __name__ == "__main__":
    print("Starting data extraction from API...")
    all_data = fetch_all_data()

    print("Saving data locally...")
    save_data_locally(all_data, filename="bike_trips.json")

    print("Inserting data into MongoDB...")
    insert_data_into_mongodb(all_data)
```
